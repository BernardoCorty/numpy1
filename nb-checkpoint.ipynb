<h1>Os 10 Jogos mais Vedidos na História dos Jogos</h1>

```python
# os 10 jogos mais vendidos na história.  ascending = legenda
top_10_vendidos = videogames[['Nome', 'Vendas Globais']].head(10).set_index('Nome').sort_values('Vendas Globais', ascending=True)
top_10_vendidos.plot(kind='barh', figsize=(11,7), grid=True, color='darkred', legend=True )
plt.title('Os 10 Jogos mais vendodos do Mundo')
plt.xlabel('Total de Vendas (em milhões de dólares)')
plt.show()
```

```python
# crio um DatFrame estruturado entre essas 2 variáveis,
# para fazer o gráfico de calor eu tenho realizor um cruzament de valores entre dados de 2 colunas
# para isso utiliza-se a função crosstab com as colunas plataforma e gênero
# e armazeno o resultado desse cruzamento na variável crosstab_vg
crosstab_vg = pd.crosstab(videogames['Plataforma'], videogames['Gênero'])
crosstab_vg.head(10)

# adiciono a coluna 'total' ao novo DataFrame
crosstab_vg['Total'] = crosstab_vg.sum(axis=1)
crosstab_vg.head(10)
```

```python
# crio um DatFrame chamado top10_platforms
top10_platforms = crosstab_vg[crosstab_vg['Total']>1000].sort_values('Total', ascending=False)

# cria o segundo DataFrame, que tb soma os jogos por gnero.
top10_final = top10_platforms.append(pd.DataFrame(top10_platforms.sum(), columns=['Total']).T, ignore_index=False)

# defino o tamanho da fonte a ser usada no gráfico
sns.set(font_scale=1)

# define o tamanho da figura

plt.figure(figsize=(18, 9))

# código para imprimir o gráfico de calor, vulgo heatmap
           
sns.heatmap(top10_final, annot=True, vmax=top10_final.loc[:'PS',:'Starategy'].values.max(), vmin=top10_final.loc[:,
:'Strategy'].values.min(), fmt='d')
            
plt.title('Os 10 Jogos mais vendodos do Mundo')
plt.xlabel('Total de Vendas (em milhões de dólares)')
plt.show()
```
